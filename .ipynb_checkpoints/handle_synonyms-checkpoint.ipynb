# Andrew list of species and synonymes

### download data

```python
#to know which encoding it is:
#with open(os.path.join(path_data, 'synonyms/trd18-splitted-synonyms_simpler.csv')) as f:
#    print(f)
df = pd.read_csv(os.path.join(path_data, 'synonyms/trd18-splitted-synonyms_simpler.csv'), 
                 encoding='latin-1' )
print(df.shape)
df.head(3)
```

### cleaning

```python
#small example
#[x.strip().strip('\"').strip('?').strip() for x in df.iloc[1]['split2'].strip('c(').strip(')').split(',')]
```

```python
def clean_syn(x):
    
    #small cleaning
    li = [i.strip().strip('\"').strip('?').strip() for i in x.strip('c(').strip(')').split(',')]
    
    #enlver ce quil y a entre parenthèse
    li = [re.sub(r'\([^)]*\)', '', i) for i in li ]
    
    #remove NA or empty syn
    li = [i for i in li if i not in [np.nan, 'NA', '', None]]
    return(list(set(li)))
```

```python
df['split2_clean'] = df['split2'].map(lambda x: clean_syn(x))
```

```python
df.head(3)
```

### remove synonyms that is already a name of a species (keeping list)

```python
#to do this we first remove syn that equal its species name
df['li_synonyms'] = df.apply(lambda x: [i for i in x['split2_clean'] if i!=x['binomial']], axis=1)
```

```python
li_species = df['binomial'].tolist()
df['li_synonyms_clean'] = df['li_synonyms'].map(lambda x: [i for i in x if i not in li_species])
```

```python
df['syn_equal_other_species'] = df.apply(lambda x: [i for i in x['li_synonyms'] if i not in x['li_synonyms_clean']],
                                              axis=1)
```

```python
#save list of species which was once used as synonymes for another species
li_species_was_syn_of_other_species = df['syn_equal_other_species'].tolist()
li_species_was_syn_of_other_species = list(set([i for x in li_species_was_syn_of_other_species for i in x if len(x)>0]))
print('There is %d species which names was use as a synonyme for another species at least once'%len(li_species_was_syn_of_other_species))
li_species_was_syn_of_other_species
pickle.dump(li_species_was_syn_of_other_species, open(os.path.join(path_data,'li_species_was_syn_of_other_species.pkl'), 'wb'))
```

```python
#for example one can see that a synonyme of Acanthophis pyrrhus is Acanthophis antarcticus, which is a name of 
#another species.
```

### remove synonymes that are in used for multiple species

```python
li_syn = df['li_synonyms_clean'].tolist()
li_syn = [i for x in li_syn for i in x if len(x)>0]
len(li_syn)
```

```python
#create a list of synonymes that appeared in several species, then remove it 
c = Counter(li_syn)
li_syn_removed = [k for k,v in c.items() if v>1]
len(li_syn_removed)
```

```python
df['li_synonyms_final'] = df['li_synonyms_clean'].map(lambda x: [i for i in x if i not in li_syn_removed])
```

```python
print(df.shape)
df
```

```python
#save andrew syn dico
dico_species_lisynandrew = dict(zip(df['binomial'], df['li_synonyms_final']))
pickle.dump(dico_species_lisynandrew, open(os.path.join(path_data,'synonyms','dico_species_lisynandrew.pkl'), 'wb'))
#save df 
df.to_csv(os.path.join(path_data, 'synonyms/df_species_syn_andrew.csv'),index=False,sep=',')
```

# language translation as synonymes (from wikipedia)

### download data

```python
f = glob.glob(os.path.join(path_data,'wikipedia','df_species_language.csv'))
if len(f)>0:
    df_wiki_syn = pd.read_csv(f[0],sep=';')
else:
    print('NOW LANGUAGE TRANSLATION YET, CAN NOT RUN BELOW')
    sys.exit()
```

```python
print(df_wiki_syn.shape)
df_wiki_syn.head(3)
```

### preprocessing

#### correct/verify info

```python
#by construction title is the english (langue in the query) language
df_wiki_syn.rename(columns={'title':'English'}, inplace=True)
```

```python
#if English variable != species, it means that it was a redirection. 
df_wiki_syn['is_redirects'] = df_wiki_syn.apply(lambda x: x['English']!=x['species'], axis=1)
```

```python
#if the englih word is another species name, then its a bad redirection and one should remove totally all these info
#first we remove the english-translation when = to its own species, then remove the line when its in the list of 
#all species
df_wiki_syn['English'] = df_wiki_syn.apply(lambda x: x['English'] if x['English']!=x['species'] else np.nan, axis=1)
df_wiki_syn['English'] = df_wiki_syn['English'].map(lambda x: x if x not in li_species else 'to_be_removed')
li_removed_lang = df_wiki_syn[df_wiki_syn['English']=='to_be_removed']['species'].tolist()
print(df_wiki_syn.shape[0])
df_wiki_syn = df_wiki_syn[df_wiki_syn['English']!='to_be_removed']
print(df_wiki_syn.shape[0])
df_wiki_syn
```

```python
#verify the redirections ANDREW :D
df_to_verify = df_wiki_syn[df_wiki_syn['is_redirects']==True]
print('There is %d redirections to be verified'%df_to_verify.shape[0])
#save df 
df_to_verify.to_csv(os.path.join(path_data, 'synonyms/df_to_verify.csv'),index=False,sep=',')
```

```python
#verify their is no two different pageid (rows) for one species
if max(list(df_wiki_syn['species'].value_counts()))!=1:
    print('ERROR: some species appeared in two wikipedia pages! check why before continuing')
    sys.exit()
```

```python
li_language = [l for l in df_wiki_syn.columns if l not in ['species','is_redirects','pageid']]
print('%d language was found in the wikipedia research'%len(li_language))
df_wiki_syn['li_lang_syn'] = df_wiki_syn.apply(lambda x: [l for l in set([x[i] for i in li_language]) if str(l)!='nan'],
                                               axis=1)
```

```python
#remove from this list the syn being the species itself
df_wiki_syn['li_lang_syn'] = df_wiki_syn.apply(lambda x: [i for i in x['li_lang_syn'] if i!=x['species']], axis=1)
```

```python
#remove from this list the syn appearing in other species synonyms as well
li_lang_syn = [i for j in df_wiki_syn['li_lang_syn'].tolist() for i in j]
len(li_lang_syn)
c = Counter(li_lang_syn)
li_multiple_times = [k for k,v in c.items() if v>1]
print(len(li_multiple_times))
df_wiki_syn['li_lang_syn'] = df_wiki_syn['li_lang_syn'].map(lambda x: [i for i in x if i not in li_multiple_times])
```

```python
#remove from this list the syn appearing as another species as well
df_wiki_syn['li_lang_syn'] = df_wiki_syn['li_lang_syn'].map(lambda x: [i for i in x if i not in li_species])
```

```python
#add number of language translation the species has
df_wiki_syn['nbr_lang'] = df_wiki_syn['li_lang_syn'].map(lambda x: len(x))
```

```python
#save language dico
dico_species_lisyn_langu = dict(zip(df_wiki_syn['species'], df_wiki_syn['li_lang_syn']))
pickle.dump(dico_species_lisyn_langu, open(os.path.join(path_data,'synonyms','dico_species_lisyn_langu.pkl'), 'wb'))
#save df 
df_wiki_syn.to_csv(os.path.join(path_data, 'synonyms/df_wiki_syn.csv'),index=False,sep=',')
```

#### small stat

```python
#produce a dictionary with the keys=species, and the values=dico with language:translation
dico_species_dico_language_transl = {}
for i in tqdm.tqdm(range(df_wiki_syn.shape[0])):
    dico = dict(zip(df_wiki_syn.iloc[i].keys(), df_wiki_syn.iloc[i].values))
    li = dico['li_lang_syn']
    species = dico['species']
    dico.pop('li_lang_syn', None)
    dico.pop('nbr_lang', None)
    dico.pop('pageid', None)
    dico.pop('species', None)
    dico.pop('title', None)
    dico = {k:v for k,v in dico.items() if v in li}
    dico_species_dico_language_transl[species] = dico
#save it
pickle.dump(dico_species_dico_language_transl, open(os.path.join(path_data,'wikipedia',
                                                                 'dico_species_dico_language_transl.pkl'), 'wb'))

#dico of language and its list of species that have a tranlation
dico_language_lispecies = defaultdict(list)
for k,v in dico_species_dico_language_transl.items():
    for k1,v1 in v.items():
        dico_language_lispecies[k1] = dico_language_lispecies[k1] + [k]
dico_language_lispecies = dict(dico_language_lispecies)
#save it
pickle.dump(dico_language_lispecies, open(os.path.join(path_data,'wikipedia',
                                                                 'dico_language_lispecies.pkl'), 'wb'))
```

# correction of synonyms

### synonyms must include no species names 

```python
# to do this we will first verify that it doe snot include the associated species (should not happen) in the list of
#synonyms
for k,v in dico_species_lisyn_.items():
    if k in v:
        print('ERROR: ', k, ' in ', v)
        sys.exit()
```

```python
#then verify all the word associated to any in li_species (that should not happen too)
for k,v in dico_species_lisyn_.items():
    for i in v:
        if i in li_species:
            print('ERROR: ', i, ' is a species')
            sys.exit()
```

### synonyms must be unique to a species

```python
#note that species synonymes coming from language might match a syn coming from adnrew list for example. lets remove
#these bad situations
li_syn = [i for j in dico_species_lisyn_.values() for i in j]
c = Counter(li_syn)
li_multiple_times = [k for k,v in c.items() if v>1]
print(len(li_multiple_times))
print(li_multiple_times)
dico_species_lisyn_final = {k:[i for i in v if i not in li_multiple_times] for k,v in dico_species_lisyn_.items()}
```

### final verification

```python
li = list(dico_species_lisyn_final.keys())
li_ = [i for j in dico_species_lisyn_final.values() for i in j]
li.extend(li_)
if len(li)!=len(set(li)):
    print('ERROR: some species/species-word are duplicates')
    sys.exit()
```
