# Download data

```python
#choose your images to analyse
image_type = 'crowdai test'
#image_type = 'crowdai train'
#image_type = 'crowdai'
#image_type = 'all'
```

```python
if image_type=='all':
    df = pd.read_csv(os.path.join(path_data,'df_all_datasource.csv'), sep=';', index_col=False)
if image_type=='crowdai':
    df = pd.read_csv(os.path.join(path_data,'df_crowdai.csv'), sep=';', index_col=False)
if image_type=='crowdai test':
    df = pd.read_csv(os.path.join(path_data,'df_crowdai_test.csv'), sep=';', index_col=False)
if image_type=='crowdai train':
    df = pd.read_csv(os.path.join(path_data,'df_crowdai_train.csv'), sep=';', index_col=False)
```

```python
print(df.shape)
df.head(3)
```

# Small stat

```python
x = len(df['species'].unique())
print('%d species have at least one image'%x)
print('We have %d images'%df.shape[0])
```

### number of images per species

```python
df_plot = df.groupby(['species'])['id'].count().reset_index()
df_plot = df_plot.sort_values('id',ascending=False)
df_plot.head(3)
```

```python
min_val_ = 0
dico_species_totalnbrimages = dict(zip(df_plot['species'].tolist(), df_plot['id'].tolist()))
donut_plot(li_labels = list(dico_species_totalnbrimages.keys()), 
           li_sizes = list(dico_species_totalnbrimages.values()),
           path = os.path.join(path_data, 'plot/nbr_images_per_species_'+image_type+'_'+str(min_val_)+'.png'),
           v=0.6, min_val=min_val_, nbr_without_explode=20, fontsize_=3, nbr_set=5)
```

```python
df_plot = df.groupby(['species','datasource'])['id'].count().reset_index()
df_plot = df_plot.sort_values('id',ascending=False)
print(df_plot.shape)
df_plot.head(3)
```

```python
min_val_=0
df_plot_ = df_plot.pivot(index='species', columns='datasource', values='id').fillna(0)
df_plot_['total_nbr_image'] = df_plot_['flickr']+df_plot_['herpmapper']+df_plot_['inaturalist']+df_plot_['snapp']
df_plot_ = df_plot_[df_plot_['total_nbr_image']>min_val_]
df_plot_ = df_plot_.sort_values('total_nbr_image',ascending=False)
df_plot_.drop(['total_nbr_image'], inplace=True, axis=1)
df_plot_.head(3)
ax = df_plot_.plot(kind='bar', stacked=True, figsize=(18.5, 10.5), fontsize=5)
plt.savefig(os.path.join(path_data, 'plot/nbr_images_per_species_bar_plot_'+image_type+'_'+str(min_val_)+'.png'), 
            dpi=300, format='png', bbox_inches='tight')
plt.title('Number of images per species')
plt.show()
```
